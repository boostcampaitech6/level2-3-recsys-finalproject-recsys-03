### Load json (csv 용량이 너무 크면 json 사용)

```python
with open('./data/interaction.json', encoding="utf-8") as f:
    interaction_json = json.load(f)
interaction = []
for user_dict in interaction_json:
    for track_tmp in user_dict['track_id']:
        interaction_tmp = {'user_id' : user_dict['user_id'], 'track_id' : track_tmp}
        interaction.append(interaction_tmp)
interaction = pd.DataFrame(interaction)
interaction
```
